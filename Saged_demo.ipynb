### Dmonstrating the Pipeline for analytics

```python
from saged import Pipeline, ResponseGenerator
import pandas as pd
from assistants import OllamaModel, GPTAgent
```

```python
ollama_base_models = ['llama3']
gpt_models = []
roles = {
    'no_role': '',
    'Donald Trump': 'Role-play Donald Trump',
    'Joe Biden': 'Role-play Joe Biden',
}

generation_dict = {}
template = ResponseGenerator.prompt_template

model_index = 0
for role in roles:
    for bm in ollama_base_models:
        model_index += 1
        ollama_agent = OllamaModel(base_model=bm,
                            system_prompt=roles[role],
                            model_name=f'{bm}{model_index}',
                            temperature=0
                            )
        generation_dict[f'{bm}_{role}'] = lambda x, agent=ollama_agent: agent.invoke(template('question_answering')(x))

    for gpt in gpt_models:
        gpt_agent = GPTAgent(gpt)
        gf_role = lambda x, agent=gpt_agent: agent.invoke(f'{roles[role]}\n{x}', temperature=0)
        # Use a default argument in the lambda to capture the current gf_role
        generation_dict[f'{gpt}_{role}'] = lambda x, role_function=gf_role: role_function(
            template('question_answering')(x))
```
